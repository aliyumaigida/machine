```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\ALIYU\\Desktop\\Datasets\\Melbourne_housing_FULL.csv')
df
```

```python
# 1. Extract from the dataset where Price column is not null into another variable

data_with_price = df[df['Price'].notna()]
data_with_price.head()
```

```python
#   2. Based on the this new variable clean the data to ensure no more NAN on the dataset by applying the following:
#         * Delete unuseful columns from the dataset
#         * Use the number of Rooms column to fill for mising number of Bedroom2
#         * Use the number of bedroom2 column to fill for the mising number of Bathroom
#         * Using this same logic, look for a column that have close relationship with another column with missing number and   
#          use it to fill for the missing values in the column
```

```python
#  * Delete unuseful columns from the dataset

columns_to_drop =['Address', 'Method', 'SellerG', 'Date', 'CouncilArea', 'Regionname', 'Propertycount']

data_clearned = data_with_price.drop(columns=columns_to_drop)
```

```python
#  * Use the number of Rooms column to fill for mising number of Bedroom2
data_clearned['Bedroom2'].fillna(data_clearned['Rooms'], inplace=True)
```

```python
#  * Use the number of bedroom2 column to fill for the mising number of Bathroom
data_clearned['Bathroom'].fillna(data_clearned['Bedroom2'], inplace=True)
```

```python
df.columns
```

```python
#   * Using this same logic, look for a column that have close relationship with another column with missing number and          
#      use it to fill for the missing values in the column



# check for missing values in columns

data_clearned.isnull().sum()
```

```python
# filling car with Bedroom2

data_clearned['Car'].fillna(data_clearned['Bedroom2'], inplace=True)
```

```python
# fill lansize and buildingArea with Car
data_clearned['Landsize'].fillna(data_clearned['Car'], inplace=True)

data_clearned['BuildingArea'].fillna(data_clearned['Car'].medi, inplace=True)
```

```python
# Fill YearBuilt with the most frequent values

data_clearned['YearBuilt'].fillna(data_clearned['YearBuilt'].mode()[0], inplace=True)
```

```python
# fill Distance  with median values

data_clearned['Distance'].fillna(data_clearned['Distance'].median(), inplace=True)
```

```python
# fill Postcode with median values

data_clearned['Postcode'].fillna(data_clearned['Postcode'].median(), inplace=True)
```

```python
# fill Lattitude with mean values

data_clearned['Lattitude'].fillna(data_clearned['Lattitude'].mean(), inplace=True)
```

```python
# fill Longtitude with mean values

data_clearned['Longtitude'].fillna(data_clearned['Longtitude'].mean(), inplace=True)
```

```python
data_clearned.isnull().sum()
```

```python
# 3. Show in a table, how Bedroom2 and landsize affect the price of house

bedroom2_landsize_price =data_clearned.groupby(['Bedroom2','Landsize']).agg({'Price':['mean', 'median']}).reset_index()

bedroom2_landsize_price
```

```python
# 4. Also show how BuildingArea and YearBuilt affect the price of house
buildingArea_Yearbuilt_price = data_clearned.groupby(['BuildingArea', 'YearBuilt']).agg({'Price':['mean', 
                                                                                                  'median']}).reset_index()

buildingArea_Yearbuilt_price
```

```python
#  5. In a tabel, show the sum and median of price of house based on Bedroom2

bedroom2_price_summary = data_clearned.groupby('Bedroom2').agg({'Price': ['sum', 'median']}).reset_index()

bedroom2_price_summary
```
