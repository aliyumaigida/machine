```python
import pandas as pd
```

```python
df1 = pd.read_csv("C:\\Users\ALIYU\\Desktop\\Datasets\\Nigeria data\\list_of_local_government_areas_of_nigeria.csv")
df1
```

```python
df2 = pd.read_csv("C:\\Users\ALIYU\\Desktop\\Datasets\\Nigeria data\\Nigeria povertyrate per local gov.csv")
df2
```

```python
df3 = pd.read_csv("C:\\Users\ALIYU\\Desktop\\Datasets\\Nigeria data\\list-states-nigeria-227j.csv")
df3
```

```python
# 1. Marge the two datasets together to form a single table.

# first merge two data

df_merge = pd.merge(df1, df2, left_on="Local Government Area", right_on="LGA/state")
df_merge

```

```python
df2['State']=df2['states']
df2.head()
```

```python
#   . Marge the three datasets together to form a single table.

df_merge2 = pd.merge(df_merge, df3, on="State")
df_merge2.head()
```

```python
df_merge2.columns
```

```python
1. # Remove any redundant columns

df_merge = df_merge2.drop(columns=['ID_0','ISO','NAME_0','SNo_x', 'SNo_y', 'Rank_y', 'Population_y'], axis=1)
df_merge.head()
```

```python
df_merge.isnull().any()
```

```python
(df_merge.isnull().sum()/len(df_merge))*100
```

```python
3. # Clean the dataset to remove every null values

df_clearn = df_merge.dropna()
df_clearn.head()
```

```python
4. # Using either groupby or pivort_table calculate how population affects poverty rate of each local government

population_poverty = df_clearn.groupby("Local Government Area").apply(lambda x: (x["2014_Popul"] * x["Poverty_Ra"]).sum() /
                                                                     x["2014_Popul"].sum()).reset_index()

population_poverty
```

```python
population_poverty.rename(columns={0 : "Weighted_poverty_rate"}, inplace=True)
population_poverty
```

```python
5# Arrange each local goverment according to the order of their poverty rate

sort_values = population_poverty.sort_values(by='Weighted_poverty_rate', ascending=False)
sort_values

#population_poverty['Weighted_poverty_rate'].sort_values(ascending=False)
```
