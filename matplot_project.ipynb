### DOB and Survey Date columns has different datetime format that is why i have to use different format to convert it to datetime

```python
from datetime import datetime
# Function to clean and convert mixed date formats
def convert_mixed_dates(date):
    if pd.isna(date) or str(date).strip() == "":
        return None  # Keep missing values as None
    date = str(date).strip()

    try:
        # First, try parsing with day-first format (handles DD-MM-YYYY)
        return pd.to_datetime(date, dayfirst=True, errors="coerce")
    except ValueError:
        try:
            # Try parsing as MM/DD/YYYY or M/D/YYYY
            return pd.to_datetime(date, format="%m/%d/%Y", errors="coerce")
        except ValueError:
            try:
                # Try parsing as YYYY-MM-DD HH:MM:SS
                return pd.to_datetime(date, format="%Y-%m-%d %H:%M:%S", errors="coerce")
            except ValueError:
                return date  # Return original value if all parsing fails

hr_data["DOB"] = hr_data["DOB"].apply(convert_mixed_dates)
hr_data["Survey Date"] = hr_data["Survey Date"].apply(convert_mixed_dates)

# Convert entire column to datetime
hr_data["DOB"] = pd.to_datetime(hr_data["DOB"], errors="coerce")
hr_data["Survey Date"] = pd.to_datetime(hr_data["Survey Date"], errors="coerce")
```

### convert from object data type to datetime, the three columns has thesame datetime format

```python
date_cols = ['StartDate', 'ExitDate', 'Training Date']
for col in date_cols:
    hr_data[col] = pd.to_datetime(hr_data[col], format="%d-%b-%y", errors='coerce')
```

## Trying to create a new column call Age by subtracting DOB from current_date

```python
current_date = datetime.today()
# Calculate age in years
hr_data["Age"] = (current_date - hr_data["DOB"]).dt.days // 365
```

## Trying to visualize the dataset

```python
import matplotlib.pyplot as plt

status_counts = hr_data['EmployeeStatus'].value_counts()
classification_counts = hr_data['EmployeeClassificationType'].value_counts()
department_counts = hr_data['DepartmentType'].value_counts()

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Employee Status
axes[0].bar(status_counts.index, status_counts.values, color="skyblue")
axes[0].set_title("Employee Status Distribution")
axes[0].set_xticks(range(len(status_counts.index)))  
axes[0].set_xticklabels(status_counts.index, rotation=45, ha="right")

# Employee Classification
axes[1].bar(classification_counts.index, classification_counts.values, color="lightcoral")
axes[1].set_title("Employee Classification Type Distribution")
axes[1].set_xticks(range(len(classification_counts.index))) 
axes[1].set_xticklabels(classification_counts.index, rotation=45, ha="right")

# Department Type
axes[2].bar(department_counts.index, department_counts.values, color="lightgreen")
axes[2].set_title("Department Distribution")
axes[2].set_xticks(range(len(department_counts.index)))  
axes[2].set_xticklabels(department_counts.index, rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
